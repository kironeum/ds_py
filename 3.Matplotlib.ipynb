## простые линейные графики

```python
# массив (вектор-строка) из 1000 значений 
# линейная последовательность между 0 и 10
x = np.linspace(0,10, 1000) # первый арг мин знач диапазона, втор арг макс знач диапазона, 3 парам размер 
x
```

```python
# нарисуем график функции 
# например, график функции синусоиды

y = np.sin(x)
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
# совмещение двух графиков в одном окне
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.show()
```

```python
# настройка графиков: цвета и вид линий

# цвета можно задавать разными способами
plt.plot(x, np.cos(x), color='blue') # по названию 
plt.plot(x, np.cos(x - 1), color='g') # первая буква цвета (rgbcmyk)
plt.plot(x, np.cos(x - 2), color='0.75') # оттенки серого в диапазоне от 0 до 1
plt.plot(x, np.cos(x - 3), color='#ffdd44') # 16-ричный код (RGB от 00 до FF )  одна F - 4 бита
plt.plot(x, np.cos(x - 4), color=(1.0, 0.2, 0.3)) # RGB-кортеж значений от 0 до 1
plt.plot(x, np.cos(x - 5), color='chartreuse') # название цвета HTML
plt.show()
```

```python
# стиль линий
первый способ
plt.plot(x, x + 5, linestyle= 'solid') # сплошная линия
plt.plot(x, x + 6, linestyle='dashed') # штриховая линия
plt.plot(x, x + 7, linestyle='dotted') # пунктирная линия
plt.plot(x, x + 8, linestyle='dashdot') # штрих-пунктирная линия
```

```python
# стиль линий
# второй способ
plt.plot(x, x + 5, linestyle='-') # сплошная линия
plt.plot(x, x + 6, linestyle='--') # штриховая линия
plt.plot(x, x + 7, linestyle=':') # пунктирная линия
plt.plot(x, x + 8, linestyle='-.') # штрих-пунктирная линия
```

```python
# объединение кодов, стиля и цвета линий в одном аргументе 
plt.plot(x, np.sin(x), '-g')
plt.plot(x, np.sin(x + 1), '--r')
plt.plot(x, np.sin(x + 2), ':k')
plt.show()
```

```python
# пределы осей
plt.plot(x, np.sin(x), '--g')

# индивидуальная настройка осей 
plt.ylim(-2, 10)
plt.xlim(0,20)
plt.show()
```

```python
# настройка обеих осей
plt.plot(x, np.sin(x), '--g')
plt.axis([0,5, -1.5, 1.5])
```

```python
# метки на графиках 

# интерактивный режим
%matplotlib notebook

# название (заголовок) графика и метки его осей
plt.plot(x, np.sin(x), '-')

plt.title("Синусоида")
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

```python
# легенда графика

plt.plot(x, np.sin(x), '-r', label='sin')
plt.plot(x, np.cos(x), '-g', label='cos')
plt.legend()

plt.show()
```

## гистограммы

```python
data = np.random.random(1000)

plt.hist(data, bins=10)
plt.show()
```

```python
data = np.random.randn(1000) # гауссово распределение (нормальное распределение)

plt.hist(data, bins=100)
plt.show()
```

```python
# настройки графика (гистограммы)
x1= np.random.normal(0, 0.8, 1000)
x2= np.random.normal(-2, 1, 1000)
x3= np.random.normal(3, 2, 1000)

plt.hist(x1, bins=50, alpha=0.5, color='r')
plt.hist(x2, bins=50, alpha=0.3, color='g', edgecolor='k')
plt.hist(x3, bins=50, alpha=0.5)
plt.show()
```

## простые диаграммы рассеивания

```python
x = np.linspace(0, 10, 50)
y = np.sin(x)

plt.plot(x,y, 'o')
plt.show()
```

```python
plt.scatter(x, y, marker='*')
```

```python
x = np.random.randn(50)
y = np.random.randn(50)

plt.scatter(x, y, marker='o')
plt.show()
```

```python
x = np.random.randn(50)
y = np.random.randn(50)

sizes = np.random.randn(50) * 100
colors = np.random.randn(50)

plt.scatter(x, y, marker='o', s=sizes, c=colors, cmap='viridis')
plt.show()
```

```python
plt.scatter(x, y, marker='o', s=sizes, c=colors, cmap='PRGn') #'cividis')
plt.show()
```

```python
plt.scatter(x, y, marker='o', s=sizes, c=colors, cmap='viridis', alpha=0.5)
plt.show()
```

## трехмерные графики

```python
from mpl_toolkits import mplot3d
```

```python
%matplotlib notebook

ax = plt.axes(projection='3d')

# имитационные данные
x = np.linspace(0, 15, 1000)
y = np.sin(x)
z = np.cos(x)

ax.plot3D(x,y,z)
x = np.random.random(100) * 15
y = np.sin(x) + np.random.randn(100) * 0.1
z = np.cos(x) + np.random.randn(100) * 0.1

ax.scatter3D(x, y, z, c=x, cmap='viridis')
plt.show
```

## Рост президентов США

```python
import pandas as pd
data = pd.read_csv('president_heights.csv')
data.head()
```

```python
data.loc[:,'height(cm)'].min()
```

```python
min_height = data.loc[:,'height(cm)'].min()
data.loc[data['height(cm)'] == min_height, 'name']
```

```python
max_height = data.loc[:,'height(cm)'].max()
data.loc[data['height(cm)'] == max_height, 'name']
```

```python
data.loc[:, 'height(cm)'].max()
```

```python
plt.hist(data.loc[:, 'height(cm)'], color='green', bins=20)

plt.title("Рост президентов США")
plt.xlabel('Рост (см)')
plt.ylabel('Количество президентов')
```
